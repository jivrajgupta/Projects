```python
import pandas as pd
from statsmodels.tsa.stattools import kpss
import numpy as np

dates = pd.date_range(start='2010-01-01', periods=90, freq='M')

np.random.seed(0)
monthly_values = np.random.randint(low=50, high=200, size=90)

df = pd.DataFrame({'Date': dates, 'Value': monthly_values})

df.set_index('Date', inplace = True)
df
```

```python
def kpss_test(time_series_data):
    # Convert the time series data to a pandas Series (if it's not already)
    time_series = pd.Series(time_series_data)
    
    # Perform KPSS test for stationarity
    result = kpss(time_series)
    
    # Extract and print test results
    print("KPSS Test Results:")
    print(f"Test Statistic: {result[0]}")
    print(f"P-Value: {result[1]}")
    print(f"Lags Used: {result[2]}")
    print(f"Critical Values:")
    for key, value in result[3].items():
        print(f"  {key}: {value}")
    
    # Check p-value and provide interpretation
    print("\nTest Conclusion:")
    if result[1] <= 0.05:
        print("Result: Data is non-stationary. Reject the null hypothesis.")
    else:
        print("Result: Data is stationary. Fail to reject the null hypothesis.")
    print("\nAdditional Information:")
    print("1. A larger Test Statistic suggests data is more non-stationary.")
    print("2. Compare Test Statistic with Critical Values for confidence levels (10%, 5%, 2.5%, 1%).")
    print("   If Test Statistic is greater than Critical Value, data is non-stationary.")
    print("   Critical Values provide a threshold above which the null hypothesis can be rejected.")

kpss_test(df.Value)
```

```python
def fit_auto_arima(time_series_data):
    model = auto_arima(time_series_data, start_p=1, start_q=1,
                       max_p=6, max_q=6,  # Assuming monthly data with seasonality 12
                       d=0, trace=True,
                       error_action='ignore',  
                       suppress_warnings=True, 
                       stepwise=True)
    
    # Get optimal ARIMA parameters
    p, d, q = model.order    
    # Print the optimal parameters
    print("Optimal ARIMA Parameters:")
    print(f"Non-seasonal Order (p, d, q): ({p}, {d}, {q})")
    
    # Return the optimal parameters
    return p, d, q
    
p, d, q= fit_auto_arima(df.Value)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Function to plot ACF and PACF for a given time series
def plot_acf_pacf(time_series_data):
    time_series = pd.Series(time_series_data)
    
    # Plot ACF
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plot_acf(time_series, lags=6, ax=plt.gca())
    plt.title('Autocorrelation Function (ACF)')
    
    # Plot PACF
    plt.subplot(1, 2, 2)
    plot_pacf(time_series, lags=6, ax=plt.gca())
    plt.title('Partial Autocorrelation Function (PACF)')
    
    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
plot_acf_pacf(df.Value)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Function to plot ACF and PACF for a given time series with user-defined significance level
def plot_acf_pacf(time_series_data, significance_level=0.05):
    # Convert the time series data to a pandas Series (if it's not already)
    time_series = pd.Series(time_series_data)
    
    # Plot ACF
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plot_acf(time_series, lags=6, alpha=significance_level, ax=plt.gca())
    plt.title('Autocorrelation Function (ACF)')
    plt.axhline(y=-1.96/np.sqrt(len(time_series)), linestyle='--', color='gray')  # Lower significant threshold
    plt.axhline(y=1.96/np.sqrt(len(time_series)), linestyle='--', color='gray')   # Upper significant threshold
    
    # Plot PACF
    plt.subplot(1, 2, 2)
    plot_pacf(time_series, lags=6, alpha=significance_level, ax=plt.gca())
    plt.title('Partial Autocorrelation Function (PACF)')
    plt.axhline(y=-1.96/np.sqrt(len(time_series)), linestyle='--', color='gray')  # Lower significant threshold
    plt.axhline(y=1.96/np.sqrt(len(time_series)), linestyle='--', color='gray')   # Upper significant threshold
    
    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
plot_acf_pacf(df.Value, 0.05)
```
